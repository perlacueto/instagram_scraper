```python
import instaloader
from datetime import datetime
from itertools import dropwhile, takewhile
import csv

import pandas as pd
```

```python
#Modificar según usuario y mes para nombre de archivo
username = 'bbvaprovincial' 
mes = 'Abril'
#modificar para seleccionar período a extraer
SINCE = datetime(2022, 4, 1) #fecha de inicio
UNTIL = datetime(2022, 4, 30) #fecha de final
```

```python
L = instaloader.Instaloader()
L.login(input("input your username: "), input("input your password: ") )
posts = instaloader.Profile.from_username(L.context, username).get_posts()
```

```python
with open('IG_'+username+mes+'.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(['type','id','username','text','date','likes','post_url','post_location','post_typename','post_mediacount','post_caption_hashtags','post_caption_mentions','post_tagged_users','post_comments','post_title'])
    k = 0  # initiate k

    for post in posts:
        postdate = post.date

        if postdate >= UNTIL:
            continue
        elif postdate <= SINCE:
            k += 1
            if k == 50: #se detiene al ver que 50 posts
                break
            else:
                continue
        else:
            print("post date: "+str(post.date))
            posturl = "https://www.instagram.com/p/"+post.shortcode
            writer.writerow(["post",post.mediaid, post.profile, post.caption, post.date,post.likes, posturl , post.location,  post.typename,
                             post.mediacount, post.caption_hashtags, post.caption_mentions, post.tagged_users,  
                             post.comments,  post.title])
            for comment in post.get_comments():
                writer.writerow(["comment",comment.id, comment.owner.username,comment.text,comment.created_at_utc,
                                 comment.likes_count,posturl])
            k = 0  # set k to 0
```
